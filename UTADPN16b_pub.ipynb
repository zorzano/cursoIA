Laboratorio de Machine Learning. Regresion neuronal.

```
Base de datos de calidad de vinos. Vamos a intentar estimar la calidad de un vino (columna 12) en base a sus parametros fisicos.

Input variables (based on physicochemical tests):\
1 - fixed acidity\
2 - volatile acidity\
3 - citric acid\
4 - residual sugar\
5 - chlorides\
6 - free sulfur dioxide\
7 - total sulfur dioxide\
8 - density\
9 - pH\
10 - sulphates\
11 - alcohol\
Output variable (based on sensory data):\
12 - quality (score between 0 and 10)
```

````python
# Importacion de libreria numpy, y bajamos el fichero con la base de datos
import numpy as np
url = "https://raw.githubusercontent.com/zorzano/cursoIA/master/WineQT.csv"
data = np.genfromtxt(url, delimiter=",", dtype=float, skip_header=1)
print(data)
````

````python
conjunto_x=data[:,:-2] # Inputs de training, quito calificacion e id
conjunto_y=data[:,-2].reshape(-1, 1) # outputs de training. Reshape necesario para que encaje en el training
x_train=conjunto_x[:1000]
y_train=conjunto_y[:1000]
# Separamos los datos de validacion
x_validate=conjunto_x[-143:]
y_validate=conjunto_y[-143:]

# Vamos a ver el shape de los datos de train y validacion
print("Shape de X: ", conjunto_x.shape)
print("Shape de Y: ",conjunto_y.shape)
print("Shape de x_train: ",x_train.shape)
print("Shape de y_train: ",y_train.shape)
print("Shape de x_validate: ",x_validate.shape)
print("Shape de y_validate: ", y_validate.shape)

````

````python
# Normalicemos hacia el cielo, monjas del Carmelo
# O sea, llevamos la media a cero, y dividimos por la desviacion estandar
mean=x_train.mean(axis=0)
x_train-=mean
std=x_train.std(axis=0)
x_train/=std

x_validate-=mean
x_validate/=std
````

````python
# Creacion de la red neuronal
from keras import models, Input
from keras import layers

model = models.Sequential()
model.add(Input(shape=(x_train.shape[1],)))
model.add(layers.Dense(60, activation='relu'))
model.add(layers.Dense(60, activation='relu'))
model.add(layers.Dense(1))
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
#mse: error cuadrático medio
#mae: mean absolute error
````

````python
# Entrenamiento del modelo
history=model.fit(x_train, y_train, epochs=50, batch_size=24, validation_data=(x_validate, y_validate))
````

````python
# Dibujamos la evolucion del entrenamiento de la red.
import matplotlib.pyplot as plt

loss=history.history['loss']
val_loss=history.history['val_loss']

epochs =  range(1, len(loss)+1)

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
````

````python
# Veamos lo que el modelo prevé para un vino que nos inventamos
import numpy as np

x_dato=[[8.0,0.3,0.63,1.6,0.081,16,29,0.99588,3.3,0.78,10.8]]
x_dato-=mean
x_dato/=std
prediccion=model.predict(x_dato)

print(prediccion)
````
