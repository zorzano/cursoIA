````python
import numpy as np
url = "https://raw.githubusercontent.com/zorzano/cursoIA/master/Credit_Card_Applications.csv"
data = np.genfromtxt(url, delimiter=",", dtype=float, skip_header=1)
print(data)
````

````python
conjunto_x=data[:, 1:-1]
# conjunto_x=data[:, [1, 2, 3, 4, 5, 6, 8, 9, 10, 11, 12, 15]]
conjunto_y=data[:,-1].reshape(-1, 1)
x_train=conjunto_x[:580]
y_train=conjunto_y[:580]
x_validate=conjunto_x[-110:]
y_validate=conjunto_y[-110:]
````

````python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(x_train, y_train)
````

````python
print("Precision training set: {:.2f}".format(tree.score(x_train, y_train)))
print("Precision del validation set: {:.2f}".format(tree.score(x_validate, y_validate)))
````

````python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", impurity=False, filled=True,class_names=['no', 'yes'])

import graphviz
with open("tree.dot") as f:
  dot_graph=f.read()
graphviz.Source(dot_graph)
# Value: las si y no de cada nodo
````
