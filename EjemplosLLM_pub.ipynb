```python
import tiktoken
enc = tiktoken.encoding_for_model("gpt-4o-mini")


token_ids = enc.encode("Estoy esprintando")
tokens = [enc.decode([token_id]) for token_id in token_ids]

print(list(zip(token_ids, tokens)))
```

```python
from openai import OpenAI
from google.colab import userdata
import numpy as np

#Acordarse de poner el secret
api_key = userdata.get("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)

texto1="Tuberculo"
texto2="Patata"
e1 = client.embeddings.create(
        model="text-embedding-ada-002",
        input=texto1
    )
e2 = client.embeddings.create(
        model="text-embedding-ada-002",
        input=texto2
    )
emb1=e1.data[0].embedding
emb2=e2.data[0].embedding
print(emb1)
print(emb2)

emb1 = np.array(emb1)
emb2 = np.array(emb2)

# Distancia euclídea
dist_euclidea = np.linalg.norm(emb1 - emb2)

# Similitud coseno
sim_cos = np.dot(emb1, emb2) / (
    np.linalg.norm(emb1) * np.linalg.norm(emb2)
)

# Distancia coseno
dist_cos = 1 - sim_cos

print("Euclidea:", dist_euclidea)
print("Similaridad del Coseno:", sim_cos)
print("Distancia del coseno:", dist_cos)
```
