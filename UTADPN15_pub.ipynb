````python
import numpy as np
url = "https://raw.githubusercontent.com/zorzano/cursoIA/master/Credit_Card_Applications.csv"
data = np.genfromtxt(url, delimiter=",", dtype=float, skip_header=1)

print(data.shape)
````

````python
conjunto_x=data[:, 1:-1]
conjunto_y=data[:,-1].reshape(-1, 1) # Ajuste a la forma correcta, de [690,] a [690,1]

x_train=conjunto_x[:500]
y_train=conjunto_y[:500]
x_validate=conjunto_x[-190:]
y_validate=conjunto_y[-190:]
````

````python
print(conjunto_x.shape)
print(conjunto_y.shape)
print(x_train.shape)
print(y_train.shape)
print(x_validate.shape)
print(y_validate.shape)
````

````python
mean=x_train.mean(axis=0)
x_train-=mean
std=x_train.std(axis=0)
x_train/=std

x_validate-=mean
x_validate/=std
print(data)
````

````python
from keras import models, Input
from keras import layers

model = models.Sequential()

model.add(Input(shape=(x_train.shape[1],)))
model.add(layers.Dense(60, activation='relu'))
model.add(layers.Dense(60, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
# Binary crossentropy: L=−log(p) si el target es 1, L=−log(1−p) si el target es 0
# accuracy: TP+TN/todo
# rmsprop. Optimizador que aumenta el aprendizaje si los gradientes son pequeños y viceversa. Es decir, suaviza.
````

````python
history=model.fit(x_train, y_train, epochs=20, batch_size=30, validation_data=(x_validate, y_validate))
````

````python
import matplotlib.pyplot as plt

loss=history.history['loss']
val_loss=history.history['val_loss']

epochs =  range(1, len(loss)+1)

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
````

````python
plt.clf()
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']

epochs =  range(1, len(loss)+1)

plt.plot(epochs, acc, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````
