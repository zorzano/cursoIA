```python
# Manera buena
import numpy as np
url = "https://raw.githubusercontent.com/zorzano/cursoIA/master/WineQT.csv"
data = np.genfromtxt(url, delimiter=",", dtype=float, skip_header=1)
print(data)
```

```python
# Quito la ultima columna porque tiene el identificador
conjunto_x=data[:, :-1]
print(data.shape, conjunto_x.shape)
x_train=conjunto_x[:1142]
x_validate=conjunto_x[-1:]
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=4)
kmeans.fit(x_train)

print("Clusters de train: {}".format(kmeans.cluster_centers_))
print(kmeans.labels_)
print(kmeans.predict(x_validate))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train[:, 2], x_train[:, 3], c=kmeans.labels_, cmap='bwr')
plt.show()
```
