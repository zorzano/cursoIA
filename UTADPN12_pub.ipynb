Laboratorio de Machine Learning. Clasificador k-nearest

```python
import numpy as np
url = "https://raw.githubusercontent.com/zorzano/cursoIA/master/Credit_Card_Applications.csv"
data = np.genfromtxt(url, delimiter=",", dtype=float, skip_header=1)
print(data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data[:, 2], data[:, 3], c=data[:, -1], cmap='bwr')
plt.show()
```

```python
# Quitamos la primera columna (id) y la ultima, approval, que va a validacion
conjunto_x=data[:, 1:-1]
x_train=conjunto_x[:680]
x_validate=conjunto_x[-10:]


conjunto_y=data[:,-1]
y_train=conjunto_y[:680]
y_validate=conjunto_y[-10:]

```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train, y_train)
```

```python
print("Predicciones validation set: {}".format(clf.predict(x_validate)))
print("Precision del validation set: {:.2f}".format(clf.score(x_validate, y_validate)))
```
