Laboratorio de Machine Learning. Regresion lineal

```
Base de datos de calidad de vinos.

Input variables (based on physicochemical tests):\
1 - fixed acidity\
2 - volatile acidity\
3 - citric acid\
4 - residual sugar\
5 - chlorides\
6 - free sulfur dioxide\
7 - total sulfur dioxide\
8 - density\
9 - pH\
10 - sulphates\
11 - alcohol\
Output variable (based on sensory data):\
12 - quality (score between 0 and 10)
```

````python
import numpy as np
url = "https://raw.githubusercontent.com/zorzano/cursoIA/master/WineQT.csv"
data = np.genfromtxt(url, delimiter=",", dtype=float, skip_header=1)
print(data)
````

````python
import matplotlib.pyplot as plt
plt.scatter(data[:, 0], data[:, -1])
plt.show()
````

````python
conjunto_x=data[:, :-2]
conjunto_y=data[:,-2].reshape(-1, 1)
x_train=conjunto_x[:1000]
y_train=conjunto_y[:1000]
x_validate=conjunto_x[-143:]
y_validate=conjunto_y[-143:]
````

````python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)

print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
````

````python
print("Score del training set: {:.2f}".format(lr.score(x_train, y_train)))
print("Score del validating set: {:.2f}".format(lr.score(x_validate, y_validate)))
# Score compara el error del modelo con la media
predict = lr.predict(x_train)
#x_train=[[1,1,2,1,2,3,4,1,2,3,3]]
print(predict)
````
